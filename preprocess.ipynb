```python
import pandas as pd
df = pd.read_parquet('./train_data.parquet')
df.head()
```

```python
df.situation[0]
```

```python
def preprocess_conversation(df):
    preprocessed_data = []

    for conv in df['conversations']:
        history = ''
        for i, turn in enumerate(conv):
            if turn['role'] == 'user' and i+1 < len(conv) and conv[i+1]['role'] == 'assistant':
                history += f'[USER] {turn['content']}'
                preprocessed_data.append({
                    'input' : history.strip(),
                    'output' : f'[Assistant] {conv[i+1]['content']}'
                })
            elif turn['role'] == 'assistant':
                history += f'[Assistant] {turn['content']}'
    return preprocessed_data
```

```python
preprocess_data = preprocess_conversation(df)
print(len(preprocess_data))
```
